```r
benfords_law_one_digit <-function(d) {
              if((d<1)|(d>9)) {stop("digit d is out of range")}
              log((d+1)/d)/log(10)
        }
benfords_law <- function(v) {
              sapply(v, benfords_law_one_digit)
        }
get_starting_digit <- function(x) {
                as.numeric(substr(x,1,1))
        }
```
